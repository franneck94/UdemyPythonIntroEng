#### Dictionaries


```python
students = {"Ben": 1, "Jan": 2, "Peter": 1, "Melissa": 4}
print(students)
```

```python
# Read element
student1 = students["Ben"]
print(student1)
```

```python
# Write element
students["Ben"] = 6
print(students)
```

```python
# Add element
students["Julia"] = 1
print(students)
```

```python
# Remove element
students.pop("Julia")
print(students)
```

#### Iterate over Dictionaries


```python
# Keys
for student_name in students:
    print(student_name)
```

```python
# Values
for student_grade in students.values():
    print(student_grade)
```

```python
# Keys and Values
for key, value in students.items():
    print(key, value)
```
