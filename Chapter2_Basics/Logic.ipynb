```python
upper_lim = 100
lower_lim = 0

my_var = int(input("Enter integer num: "))
```

```python
if my_var < upper_lim and my_var > lower_lim:  # (0, 100)
    print("yes")
else:
    print("no")

if not (my_var < upper_lim and my_var > lower_lim):  # (-inf, 0] or [100, inf)
    print("yes")
else:
    print("no")

if my_var >= upper_lim or my_var <= lower_lim:  # (-inf, 0] or [100, inf)
    print("yes")
else:
    print("no")
```
